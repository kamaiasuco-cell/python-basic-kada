```python

class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age


    def check_adult(self):
        if self.age >= 20:
            print(self.name, "は大人です。")
        else:
            print(self.name, "は大人ではありません。")



human1 = Human("太郎", 25)
human2 = Human("花子", 18)
human3 = Human("次郎", 20)


human_list = [human1, human2, human3]


for human in human_list:
    human.check_adult()
```
